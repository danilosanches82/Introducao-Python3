### 1) Importando pandas

### 2) Leitura dos dados

```python
data_beers = pd.read_csv('beers.csv')
data_brew = pd.read_csv('breweries.csv')
```

```python
data_brew
```

### 3) União das duas bases de dados

```python
##### Padronizar a coluna brewery_id em breweries.csv
data_brew.columns = ['brewery_id', 'name', 'city', 'state']
```

```python
##### Cria um único DataFrame com as duas bases
df = pd.merge(data_beers, data_brew, on=['brewery_id'])
```
